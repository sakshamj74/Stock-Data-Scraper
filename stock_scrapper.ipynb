```python
from bs4 import BeautifulSoup
import requests
import time
import datetime
import pandas as pd
```

```python
f=[]
z=[]
url="https://www.moneycontrol.com/stocks/marketstats/indexcomp.php?optex=NSE&opttopic=indexcomp&index=9"
response = requests.get(url)
soup = BeautifulSoup(response.text,'html.parser')
i=soup.find("table",class_="tbldata14 bdrtpg")
for k in i.findAll('td',class_="brdrgtgry"):
    if k is not None and k.get('align')=='right':
            z.append(k.text)
    for l in k.findAll('a'):
        if l.find('b') is not None:
            f.append(l.find('b').text)
```

```python
name=[]
industry=[]
Lastprice=[]	
Change=[]	
p_Chg=[]	
Mkt_Cap=[]
for i in range(0,len(f)-1,2):
    name.append(f[i])
    industry.append(f[i+1])
for i in range(0,len(z)-3,4):
        Lastprice.append(z[i])
        Change.append(z[i+1])
        p_Chg.append(z[i+2])
        Mkt_Cap.append(z[i+3])
```

```python
stock=pd.DataFrame()
stock['name']=name
stock['industry']=industry
stock['Lastprice']=Lastprice	
stock['Change']=Change	
stock['%chg']=p_Chg
stock['Mkt_Cap']=Mkt_Cap
```

```python
#Function for generating alert when %chg is greater than 2
#When %chg is greater than 2 i am printing alert along with head of the stock table
#otherwise %chg < 2  is printed
def alert(s,t):
    t['d']= pd.to_numeric(s['%chg']).abs()-pd.to_numeric(t['%chg']).abs()
    if t['d'][t['d'] > 2.0 ].count()>0:
        print('########################################################')
        print("ALERT ALERT ALERT ALERT ALERT ALERT ALERT ALERT")
        print('%chg > 2')
        print('#########################################################')
        return 1
    return 0
```

```python
now = datetime.datetime.now()
now_plus = now + datetime.timedelta(minutes = 20)
z=[]
temp=pd.DataFrame()
del p_Chg[:]
url="https://www.moneycontrol.com/stocks/marketstats/indexcomp.php?optex=NSE&opttopic=indexcomp&index=9"
while(datetime.datetime.now()<now_plus):
    z=[]
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'html.parser')
    i=soup.find("table",class_="tbldata14 bdrtpg")
    for k in i.findAll('td',class_="brdrgtgry"):
        if k is not None and k.get('align')=='right':
            z.append(k.text)
    for i in range(0,len(z)-3,4):
        p_Chg.append(z[i+2])
    temp['%chg']=p_Chg
    if alert(stock,temp):
        print(stock.head())
    stock['%chg']=p_Chg
    print("%Chg < 2")
    del p_Chg[:]
    time.sleep(30)
```
